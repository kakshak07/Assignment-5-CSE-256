```python
import matplotlib.pyplot as plt
epochs=[1,2,3,4,5]
recall=[0.225,0.383,0.415,0.423,0.427]
f_1score=[0.223,0.378,0.409,0.417,0.420]
fig=plt.figure(figsize=(10,8))
precision=[0.212,0.372,0.401,0.410,0.413]
plt.plot(epochs,precision,label="Precision")
plt.plot(epochs,recall,label="Recall")
plt.plot(epochs,f_1score,label="F-1 Score")
x_labels = ['1', '2', '3','4','5']
plt.xticks(ticks=epochs, labels=x_labels)
plt.xlabel("Epochs")
plt.ylabel("Scores")
plt.legend()
plt.title("Results of IBM 1 model")
plt.show()
```
